```python
jari_jari_tabung = float(input("silahkan masukan jari jari tabung: "))
tinggi_tabung = float(input("silahkan masukan tinggi tabung: "))

π = 3.14

volume = π * jari_jari_tabung**2 * tinggi_tabung

print(f"volume tabung adalah {volume} cm^3")
```
